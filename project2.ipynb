```python
# Let's first check which columns have missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Identify and address missing values
# For numerical features
numerical_cols = ['INCOME', 'CREDIT_SCORE', 'ANNUAL_MILEAGE']
for col in numerical_cols:
    # Check for missing values
    if df[col].isnull().any():

        # imputing with median
        median_value = df[col].median()
        df[col].fillna(median_value, inplace=True)

# For categorical features
categorical_cols = ['GENDER', 'EDUCATION', 'VEHICLE_TYPE']
for col in categorical_cols:
    # Check for missing values
    if df[col].isnull().any():
        # imputation with mode
        mode_value = df[col].mode()[0]
        df[col].fillna(mode_value, inplace=True)


# Identify and correct inconsistencies in categorical values
#  for 'GENDER'
df['GENDER'] = df['GENDER'].str.lower()
# Check unique values again to ensure consistency
print('GENDER', df['GENDER'].unique())

# Checking for inconsistency in 'EDUCATION' column
df['EDUCATION'] = df['EDUCATION'].str.lower()
print('Unique values in EDUCATION:', df['EDUCATION'].unique())

# Checking for inconsistency in 'VEHICLE_TYPE' column
df['VEHICLE_TYPE'] = df['VEHICLE_TYPE'].str.lower()
print('Unique values in VEHICLE_TYPE:', df['VEHICLE_TYPE'].unique())



# Produce univariate visuals

# for 'AGE'
sns.histplot(df['AGE'], kde=True)
plt.title('Distribution of Age')
plt.show()


# 'DRIVING_EXPERIENCE' (Categorical)
sns.countplot(x='DRIVING_EXPERIENCE', data=df)
plt.title('Distribution of Driving Experience')
plt.show()

# 'RACE' (Categorical)
sns.countplot(x='RACE', data=df)
plt.title('Distribution of Race')
plt.show()

# 'EDUCATION' (Categorical)
sns.countplot(x='EDUCATION', data=df)
plt.title('Distribution of Education')
plt.show()

# 'INCOME' (Categorical)
sns.countplot(x='INCOME', data=df)
plt.title('Distribution of Income')
plt.show()

# 'VEHICLE_OWNERSHIP' (Categorical)
sns.countplot(x='VEHICLE_OWNERSHIP', data=df)
plt.title('Distribution of Vehicle Ownership')
plt.show()

# 'VEHICLE_YEAR' (Categorical)
sns.countplot(x='VEHICLE_YEAR', data=df)
plt.title('Distribution of Vehicle Year')
plt.xticks(rotation=45)
plt.show()

# 'MARRIED' (Categorical)
sns.countplot(x='MARRIED', data=df)
plt.title('Distribution of Marital Status')
plt.show()

# 'CHILDREN' (Categorical)
sns.countplot(x='CHILDREN', data=df)
plt.title('Distribution of Children')
plt.show()

# 'POSTAL_CODE' (Categorical)
sns.countplot(x='POSTAL_CODE', data=df)
plt.title('Distribution of Postal Code')
plt.show()

# 'VEHICLE_TYPE' (Categorical)
sns.countplot(x='VEHICLE_TYPE', data=df)
plt.title('Distribution of Vehicle Type')
plt.show()

# 'SPEEDING_VIOLATIONS' (Numerical)
sns.histplot(df['SPEEDING_VIOLATIONS'], kde=True)
plt.title('Distribution of Speeding Violations')
plt.show()

# 'DUIS' (Numerical)
sns.histplot(df['DUIS'], kde=True)
plt.title('Distribution of DUIs')
plt.show()

# 'PAST_ACCIDENTS' (Numerical)
sns.histplot(df['PAST_ACCIDENTS'], kde=True)
plt.title('Distribution of Past Accidents')
plt.show()

# 'OUTCOME' (Categorical)
sns.countplot(x='OUTCOME', data=df)
plt.title('Distribution of Outcome')
plt.show()


#  Identify outliers
# Use box plot to identify outliers in numerical features

#  for 'CREDIT_SCORE'
sns.boxplot(x=df['CREDIT_SCORE'])
plt.title('Boxplot of Credit Score')
plt.show()


```
